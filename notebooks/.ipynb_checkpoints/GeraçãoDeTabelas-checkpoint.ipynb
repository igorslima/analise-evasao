```python
def gerar_graficos(dados, coluna, tipo_grafico, titulo, ano):
    data = dados[coluna].value_counts()
    data.plot(kind=tipo_grafico, title=titulo,figsize=(10, 9))
    plt.savefig('imagens' +'/'+ ano +"/" + titulo)
```

```python
def gerar_todos_os_graficos(data_frame, ano):
    for coluna in data_frame:
        gerar_graficos(data_frame,coluna, 'barh', coluna,ano)
```
